```python
#Importing all required Libraries
import cv2
from matplotlib import pyplot as plt
from PIL import Image
from IPython.display import display 
%matplotlib inline
```

```python
#Loading and resizing of Image of a bird
original_image = cv2.imread("Bird.jpg")
original_image=cv2.resize(original_image,(600,500))
converted_image=cv2.cvtColor(original_image,cv2.COLOR_BGR2RGB)
display("Original Bird Image")
plt.imshow(converted_image)
plt.show()
```

```python
# Converting the image to gray scale
gray_scale_image=cv2.cvtColor(converted_image,cv2.COLOR_BGR2GRAY)
display("Gray Scale Image")
display(Image.fromarray(gray_scale_image))
```

```python
# Inversion of Gray-scale image to a negative image
inverted_image=cv2.bitwise_not(gray_scale_image)
display("Inverted Image")
display(Image.fromarray(inverted_image))
```

```python
# To remove the noise from the images or to reduce the details from the images, we make use of Gaussian blurring.
blurred_image = cv2.GaussianBlur(inverted_image, (15, 15),0, 0)
display("Blurred Image")
display(Image.fromarray(blurred_image))
```

```python
# Inversion of blurred image
inverted_blurred_image=255-blurred_image
pencil_sketch=cv2.divide(gray_scale_image, inverted_blurred_image, scale=250.0)
display("Final Pencil Sketch of the Bird")
display(Image.fromarray(pencil_sketch))
```

```python
# Comparison of original and Pencil sketch
plt.imshow(converted_image)
display(Image.fromarray(pencil_sketch))
```
