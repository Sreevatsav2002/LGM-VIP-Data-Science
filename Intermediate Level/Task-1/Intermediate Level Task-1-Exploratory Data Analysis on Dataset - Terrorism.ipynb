# Modifying Data Columns

```python
col = dataset.columns
col
```

```python
dataset = pd.DataFrame(dataset, dataset.rename(columns={ 'success': 'Success', 'location': 'Location', 'iyear':'Year','imonth':'Month','extended':'Extended','iday':'Day','country_txt':'Country',
                   'country': 'Country_code', 'provstate':'State','region_txt':'Region','attacktype1_txt':'Attack_Type','target1':'Target',
                   'nkill':'Total_Killed','nwound':'Total_Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                   'weaptype1_txt':'Weapon_Type','motive':'Motive','city':'City','latitude':'Latitude','longitude':'Longitude'},inplace=True))
```

```python
dataset =  dataset[['Year','Month','Extended','Day','Country','State','Region','City','Latitude','Longitude','Location','Attack_Type','Total_Killed','Total_Wounded','Target','Summary','Group','Target_type','Weapon_Type', 'Motive', 'Success']]
```

### **Data Visualization and Analysis**

```python
dataset[['Group','Total_Killed']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('Total_Killed',ascending=False).head(10).plot(kind='bar',color='brown',figsize=(20,10))
plt.title("Top 10 Terrorist Attack Groups",fontsize=20)
plt.xlabel("Terrorist Groups",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(25,10))
sns.barplot(dataset['Attack_Type'].value_counts()[:20].index,dataset['Attack_Type'].value_counts()[:20].values,palette='mako')
plt.xlabel("Attack Type",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.title("Most used Attack Types",fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(20,5))
sns.countplot(dataset["Target_type"],order=dataset['Target_type'].value_counts().index,palette="mako");
plt.xticks(rotation=90)
plt.xlabel("Target type",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.title("Most Attacked Target types",fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(20,5))
sns.barplot(dataset['Country'].value_counts()[:20].index,dataset['Country'].value_counts()[:20].values,palette='mako')
plt.title('Top 20 Most Countries Attacked', fontsize = 20)
plt.xlabel('Country', fontsize = 15)
plt.ylabel('Number of attacks', fontsize = 15)
plt.xticks(rotation = 50)
plt.show()
```

```python
plt.subplots(figsize=(20,5))
sns.barplot(dataset['City'].value_counts()[:20].index,dataset['City'].value_counts()[:20].values,palette='mako')
plt.xlabel("Cities",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.title("Top 20 Most Cities Attacked",fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(25,10))
sns.barplot(dataset['Region'].value_counts()[:20].index,dataset['Region'].value_counts()[:20].values,palette='mako')
plt.xlabel("Region",fontsize=15)
plt.ylabel("Number of attacks",fontsize=15)
plt.title("Top Regions Attacked",fontsize=20)
plt.show()
```
